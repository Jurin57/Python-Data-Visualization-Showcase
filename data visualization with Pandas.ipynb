```python
import pandas as pd
import matplotlib.pyplot as plot

iris = pd.read_csv ("C:\\Users\\jyy20\\Desktop\\fallB\\python\\week7 visualization\\iris_dataset.csv")
wine = pd.read_csv ("C:\\Users\\jyy20\\Desktop\\fallB\\python\\week7 visualization\\wine.csv")

iris.head()
```

```python
# create the scatter plot
iris.plot.scatter(x = "sepal length (cm)", y = "sepal width (cm)", title = "Iris Data Set")
```

```python
# line chart
iris.drop(["target"], axis = 1).plot.line(title = "Iris Data Set")
```

```python
# histogram
wine ["points"].plot.hist()
```

```python
# multiple histograms
iris.plot.hist(subplots = True, layout = (1,4), figsize = (20,10), bins = 30)
```

```python
# bar chart (vertical)
# also have to calculate the freqency 
wine["points"].value_counts().sort_index().plot.bar()
```

```python
# bar chart (horizontal)
wine["points"].value_counts().sort_index().plot.barh()
```

```python
# plot country by average price of wine based on the dataset
# create a group
df = wine.groupby("country")
# get the price column
df = df["price"]
# calculate the average price per group(country)
df = df.mean()
# sort the values in the descending order
df = df.sort_values(ascending = False)

# plot the bar chart
df.plot.bar()
```

```python
# only look at the top countries
wine.groupby("country")["price"].mean().sort_values(ascending = False)[:5].plot.bar()
```

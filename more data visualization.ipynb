```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plot
iris = pd.read_csv ("C:\\Users\\jyy20\\Desktop\\fallB\\python\\week7 visualization\\iris_dataset.csv")
wine = pd.read_csv ("C:\\Users\\jyy20\\Desktop\\fallB\\python\\week7 visualization\\wine.csv")
```

```python
# beyond measures od central tendancy-- mean, median, mode
# five number summary-- minimum value(0th percentile), first quartile(25th percentile), median(50th percentile), third quarter(75th percentile, maximum value(100th percentile)

# box plot
# fileter on some conditions
condition1 = wine['points'] >= 95
condition2 = wine["price"] < 1000

# create the new dataset
df = wine[(condition1 & condition2)]

# create the box plot
sns.boxplot("points", "price", data = df)
```

```python
# heat map
# get the correlation matrix
corr = iris.corr()

# create the figure and axis
figure, axis = plot.subplots()

# create the heatmap
im = axis.imshow (corr.values)

# set the labels
axis.set_xticks (np.arange (len (corr.columns)))
axis.set_yticks (np.arange (len (corr.columns)))
axis.set_xticklabels (corr.columns)
axis.set_yticklabels (corr.columns)

# rotate the tick labels and set their alignment
plot.setp (axis.get_xticklabels(), rotation = 45, ha = "right", rotation_mode = "anchor")

# loop over data dimensions and create text annotations
for i in range (len (corr.columns)):
    for j in range (len (corr.columns)):
        # get the text at position [j, i] with the value at corr.iloc [i, j] rounded to 2 decimal places.
        text = axis.text(j, i, np.around (corr.iloc[i, j], decimals = 2), ha = "center", va = "center", color = "black")
```
